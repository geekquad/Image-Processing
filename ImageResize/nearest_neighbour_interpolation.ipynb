# Nearest-neighbor interpolation 

```python
import cv2
from matplotlib import pyplot as plot
import numpy as nm
```

```python
img =cv2.imread("apple.jpg")[:,:,::-1]
plot.imshow(img)
plot.title("Original Image")
```

#### Dimension Of Orignal Image

```python
print(img.shape)
```

### Applying Nearest-neighbor interpolation 

```python
near_img = cv2.resize(img,None, fx = 10, fy = 10, interpolation = cv2.INTER_NEAREST)
plot.imshow(near_img)
plot.title("Nearest Neighbor Image")
```

#### Dimension after applying Nearest-neighbor interpolation

```python
print(near_img.shape)
```
