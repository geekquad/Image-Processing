## Morphological Transformations

```python

import cv2
import numpy as np
from matplotlib import pyplot as plt
```

## Erosion
Erosion shrinkens the image pixels.
cv2.erode() method is used to perform erosion

#### Syntax : cv2.erode(src, kernel,iterations = 1)
##### src : Image on which erosion has to be performed
##### kernel : A structuring element used for erosion. If element = Mat(), a 3 x 3 rectangular structuring element is used.
##### iteration : Number of times erosion is performed

### Structuring Element 
A structuring element is a matrix that identifies the pixel in the image being processed and defines the neighborhood used in the processing of each pixel. You typically choose a structuring element the same size and shape as the objects you want to process in the input image




```python
image = cv2.imread('./assets/circles.png')
#Original Image
plt.subplot(1 , 2 , 1)
plt.title('Original Image')
plt.imshow(image)
#Kernel 
kernel = np.ones((5,5) , np.uint8)
# Erosion 
eroded_img = cv2.erode(image, kernel , iterations = 1)
plt.subplot(1,2,2)
plt.title('Eroded Image')
plt.imshow(eroded_img)

```

## Dilation
Dilation adds pixels to the boundaries of objects in an image.
cv2.dilate() method is used to perform Dilation.
The syntax is similar to erosion

```python
#Original Image
plt.subplot(1 , 2 , 1)
plt.title('Original Image')
plt.imshow(image)
#Kernel 
kernel = np.ones((5,5) , np.uint8)
# Dilation 
dilated_img = cv2.dilate(image, kernel , iterations = 1)
plt.subplot(1,2,2)
plt.title('Dilated Image')
plt.imshow(dilated_img)
```

## Opening
Opening is Erosion followed by dilation . 

```python
image_opening = cv2.imread('./assets/j.png')

#Original Image
plt.subplot(1 , 2 , 1)
plt.title('Original Image')
plt.imshow(image_opening)
#Kernel 
kernel = np.ones((5,5) , np.uint8)
#Dilation 
opening = cv2.morphologyEx(image_opening, cv2.MORPH_OPEN, kernel)
plt.subplot(1,2,2)
plt.title('Opening')
plt.imshow(opening)
```

## Closing
Closing is opposite of Opening . It performs Dilation after erosion

```python
image_closing = cv2.imread('./assets/closing.png')
#Original Image

plt.subplot(1 , 2 , 1)
plt.title('Original Image')
plt.imshow(image_closing)
#Kernel 
kernel = np.ones((5,5) , np.uint8)
#Dilation 
closing = cv2.morphologyEx(image_closing, cv2.MORPH_CLOSE, kernel)
plt.subplot(1,2,2)
plt.title('Closing')
plt.imshow(closing)
```

## Morphological Gradient
It is the difference between Dilation and Erosion of image

```python
#Original Image
plt.subplot(1 , 2 , 1)
plt.title('Original Image')
plt.imshow(image)
#Kernel 
kernel = np.ones((5,5) , np.uint8)
#Dilation 
gradient = cv2.morphologyEx(image, cv2.MORPH_GRADIENT, kernel)
plt.subplot(1,2,2)
plt.title('Gradient')
plt.imshow(gradient)
```

## Top Hat
It is the difference between input image and opening of image

```python
image_tophat = cv2.imread('./assets/tophat.png')
#Original Image
plt.subplot(1 , 2 , 1)
plt.title('Original Image')
plt.imshow(image_tophat)
#Kernel 
kernel = np.ones((10,10) , np.uint8)
#Dilation 
tophat = cv2.morphologyEx(image_tophat, cv2.MORPH_TOPHAT, kernel)
plt.subplot(1,2,2)
plt.title('Top Hat')
plt.imshow(tophat)
```

## Black Hat
It is the difference between closing image and input image

```python
image_blackhat = cv2.imread('./assets/blackhat.png')
#Original Image
plt.subplot(1 , 2 , 1)
plt.title('Original Image')
plt.imshow(image_blackhat)
#Kernel 
kernel = np.ones((4,4) , np.uint8)
#Dilation 
blackhat = cv2.morphologyEx(image_blackhat, cv2.MORPH_BLACKHAT, kernel)
plt.subplot(1,2,2)
plt.title('Black Hat')
plt.imshow(blackhat)
```
