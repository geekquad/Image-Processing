# Shi Tomasi Corner Detection and Good Features to Track

```python
import cv2
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## For flat images

```python
flat_chessboard = cv2.imread('flat_chessboard.png')
flat_chessboard = cv2.cvtColor(flat_chessboard, cv2.COLOR_BGR2RGB) 
plt.imshow(flat_chessboard)
```

```python
gray_flat_chessboard = cv2.cvtColor(flat_chessboard, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_flat_chessboard, cmap='gray')
```

```python
corners_tracked = cv2.goodFeaturesToTrack(image=gray_flat_chessboard, maxCorners=64, qualityLevel=0.01, minDistance=10) # corner finding function
# PARAMETERS FOR cv2.goodFeaturesToTrack:
# image = takes the numpy array of the grayscale image (int8, float32)
# maxCorners = maximum number of corners to return
# qualityLevel = minimal accepted quality of image corners
# minDistance = minimum possible Euclidean distance between the returned corners

corners_tracked = np.int0(corners_tracked) # converts floating point values of corners_detected to int values

for i in corners_tracked:
    x, y = i.ravel() # flattens the array into a 1-D array in-place
    cv2.circle(flat_chessboard, (x, y), 10, (255, 0, 0), -1) # drawing a circle wherever corners were detected

plt.imshow(flat_chessboard)
```

## For real world images

```python
real_chess = cv2.imread('real_ch.jpg')
real_chess = cv2.cvtColor(real_chess, cv2.COLOR_BGR2RGB)
plt.imshow(real_chess)
```

```python
gray_chessboard = cv2.cvtColor(real_chess, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_chessboard, cmap='gray')
```

```python
corners_tracked = cv2.goodFeaturesToTrack(gray_chessboard, 80, 0.01, 10)

corners = np.int0(corners_tracked)

for i in corners_tracked:
    x,y = i.ravel()
    cv2.circle(real_chess, (x,y), 4, (255, 0, 0), -1)

plt.imshow(real_chess)
```
