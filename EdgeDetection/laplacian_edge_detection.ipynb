```python
#Laplacian Edge Detection
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("./cube.png")
plt.imshow(img)
plt.show()
```

```python
#Converting to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Defining the positive and negative laplacian kernels
positive_laplacian_kernel = np.array(([[0, 1, 0], [1, -4, 1], [0, 1, 0]]), np.float32)
negative_laplacian_kernel = np.array(([[0, -1, 0], [-1, 4, -1], [0, -1, 0]]), np.float32)
```

```python
# Convolve the input image with positive and negative laplacian kernels
positive_laplacian = cv2.filter2D(gray, -1, positive_laplacian_kernel)
negative_laplacian = cv2.filter2D(gray, -1, negative_laplacian_kernel)

# Scaling back to range [0 to 255]
positive_laplacian = (positive_laplacian * 255).astype(np.uint8)
negative_laplacian = (negative_laplacian * 255).astype(np.uint8)
```

```python
# Positive laplacian kernel taking out the outward edges   
plt.imshow(positive_laplacian,cmap='gray')
plt.show()
```

```python
# Negative laplacian kernel taking out the inward edges.   
plt.imshow(negative_laplacian,cmap='gray')
plt.show()
```
