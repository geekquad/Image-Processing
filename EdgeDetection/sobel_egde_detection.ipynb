# Sobel Edge Detection

```python
import cv2
from matplotlib import pyplot as plot
import numpy as nm
```

```python
img =cv2.imread("House.jfif")
gray_image=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plot.imshow(img)
plot.title("Original Image")
```

### Sobel Edge Detection in X-Direction:

```python
sobel_img_x = cv2.Sobel(gray_image,cv2.CV_32F,1,0,ksize=1) 
plot.imshow(sobel_img_x)
plot.title("Sobel Image X-Direction")
```

### Sobel Edge Detection in Y-Direction:

```python
sobel_img_y = cv2.Sobel(gray_image,cv2.CV_32F,0,1,ksize=1) 
plot.imshow(sobel_img_y)
plot.title("Sobel Image Y-Direction")
```
