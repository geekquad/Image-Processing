# Probabilistic Hough Line Transformation

```python
import cv2
from matplotlib import pyplot as plot
import numpy as nm
```

```python
img_hough = cv2.imread("soduku_img.jpg") #Reading the image
gray_img = cv2.cvtColor(img_hough,cv2.COLOR_BGR2GRAY)
canny_img = cv2.Canny(gray_img,50,100,apertureSize = 3) #Applying cannny edge detection
```

```python
plot.imshow(img_hough)
plot.title("Original Image")
```

```python
minimum_line_length=80
maximum_line_gap=10
Houghlines_PL = cv2.HoughLinesP(canny_img,1,nm.pi/180,1,minimum_line_length,maximum_line_gap) #Applying probablistic Hough lines
```

```python
for x1,y1,x2,y2 in Houghlines_PL[0]:
    cv2.line(img_hough,(x1,y1),(x2,y2),(0,255,0),2) #Drawing the detected lines
```

```python
plot.imshow(img_hough)
plot.title("Probabilistic Hough Line")
```
