# Hough Circle Transform

```python
import cv2
from matplotlib import pyplot as plot
import numpy as nm
```

```python
img_circles=cv2.imread("ball_ref_img.jpg")
gray_circle=cv2.cvtColor(img_circles,cv2.COLOR_BGR2GRAY) #Applying gray filter
```

```python
plot.imshow(img_circles)
plot.title("Original Image")
```

```python
circles = cv2.HoughCircles(gray_circle, cv2.HOUGH_GRADIENT, 1, 40, param1=40, param2=40, minRadius=10, maxRadius=60) # Applying Hough Circles
if circles is not None:
    circle1 = nm.round(circles[0, :]).astype("int")
    for (x, y, r) in circle1:
        cv2.circle(img_circles, (x, y), r, (255, 0, 0), 3) #Drawing the detected circles
```

```python
plot.imshow(img_circles)
plot.title("Hough Circles")
```
